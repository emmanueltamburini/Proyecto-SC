# Detección y recortamiento de caras en data

```python
import cv2
import matplotlib.pyplot as plt
import glob

#Clasificadores en cascada de OpenCV
Clasificador = cv2.CascadeClassifier("Clasificadores\haarcascade_frontalface_default.xml")
Clasificador2 = cv2.CascadeClassifier("Clasificadores\haarcascade_frontalface_alt2.xml")
Clasificador3 = cv2.CascadeClassifier("Clasificadores\haarcascade_frontalface_alt.xml")
Clasificador4 = cv2.CascadeClassifier("Clasificadores\haarcascade_frontalface_alt_tree.xml")

#Lista de las carpetas dentro de Data
Polaridad = glob.glob("Data\*")

#Contador para nombramiento de imágenes nueva 
i=1

for X in Polaridad:
    print("*******%s*******"%X)
    for Participantes in glob.glob("%s\\*" %X):
        #Abrir la imagen
        imagen = cv2.imread(Participantes,cv2.IMREAD_GRAYSCALE) 


        #Detección del rostro con los 4 clasificadores
        face = Clasificador.detectMultiScale(imagen, scaleFactor=1.1, minNeighbors=10, minSize=(5, 5), flags=cv2.CASCADE_SCALE_IMAGE)
        face_two = Clasificador2.detectMultiScale(imagen, scaleFactor=1.1, minNeighbors=10, minSize=(5, 5), flags=cv2.CASCADE_SCALE_IMAGE)
        face_three = Clasificador3.detectMultiScale(imagen, scaleFactor=1.1, minNeighbors=10, minSize=(5, 5), flags=cv2.CASCADE_SCALE_IMAGE)
        face_four = Clasificador4.detectMultiScale(imagen, scaleFactor=1.1, minNeighbors=10, minSize=(5, 5), flags=cv2.CASCADE_SCALE_IMAGE)

        #Tomar el primer rostro que consiga
        if len(face) == 1:
            print("Primer clasificador")
            facefeatures = face
        elif len(face_two) == 1:
            print("Segundo clasificador")
            facefeatures = face_two
        elif len(face_three) == 1:
            print("Tercer clasificador")
            facefeatures = face_three
        elif len(face_four) == 1:
            print("Cuarto clasificador")
            facefeatures = face_four
        else:
            print("Fallo clasificador")
            facefeatures = ""
        
        #Cortar y guardar el rostro
        for (x, y, w, h) in facefeatures: 
            imagen = imagen[y:y+h, x:x+w] 

            try:
                #Redimension para el modelo
                out = cv2.resize(imagen, (350, 350)) 
                #Guardar la imagen
                if X == "Data\\Negativos":
                    cv2.imwrite("Data_Tratada\\Negativos\\Data" +str(i)+".png", out) 
                    
                elif X == "Data\\Positivos":
                    cv2.imwrite("Data_Tratada\\Positivos\\Data" +str(i)+".png", out) 
                    
                else:
                    cv2.imwrite("Data_Tratada\\Neutral\\Data" +str(i)+".png", out)
                
                i=i+1
            except:
                pass #Si consigue un error 
```

# Manejo de data procesada a un data frame

```python
import cv2
import matplotlib.pyplot as plt
import glob
import numpy as np
import pandas as pd
import os.path

plt.rcParams['image.cmap'] = 'gray'

if(os.path.isfile("Alumnos.csv")):
    Alumnos=pd.read_csv("Alumnos.csv")
    print("Archivos cargados")
    
else:
    #Lista de las carpetas dentro de Data
    Polaridad = glob.glob("Data_Tratada\*")

    #Contador de alumnos
    i=0

    #Calidad de la imagen (modificar si se modifica la imagen)
    Calidad=350*350

    #Label para los pixeles de imagen
    Label=[]
    for x in range(Calidad):
        Label.append("Pixel"+str(x+1))

    #Label de la clasificación de imágenes
    Label_img=[]

    #Data frame donde se guardará la data resultante
    Alumnos = pd.DataFrame(columns=Label)

    #Introducir todas las imágenes al data frame 
    for X in Polaridad:
        print("*******%s*******"%X)
        for Participantes in glob.glob("%s\\*" %X):
            #Abrir la imagen
            if X=="Data_Tratada\\Negativos":
                Label_img.append(0)
            elif X=="Data_Tratada\\Positivos":
                Label_img.append(2)
            else:
                Label_img.append(1)

            img = cv2.imread(Participantes, cv2.IMREAD_GRAYSCALE)
            img=img/255
            img=img.reshape(Calidad)
            Alumnos.loc[i]=img
            i+=1
            print("Imagen procesada")

    #Introducir las etiquetas al Data frame
    Alumnos["Label"] = Label_img

else:
    print("archivo ")
```
